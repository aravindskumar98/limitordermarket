```python
import pandas as pd
import numpy as np
df = pd.read_csv('dummyData.csv')
```

```python
df.head()
```

```python
## drop unnecessary columns
df.drop(columns=["symbol","series","orno","acttype","tmc","cmc","mkt","new_cat"],inplace=True)
```

```python
df.head()
```

```python
## renamining the identifier column
df.rename(columns={"Unnamed: 0": "id"},inplace = True)
```

```python
df.head()
```

```python
## function to convert time stamp to an interger --> This allows for easier comparisons
def timeChange(rows):
  timeStr = rows.time
  time = np.array(timeStr.split(':'),dtype=int)
  converter = np.array([3600,60,1])
  rows.time = np.sum(time*converter)
  return rows
```

```python
## changing time stamp to int
df = df.apply(timeChange, axis = 1)
```

```python
df.head()
```

```python
##label encoding buy and sell -- Buy : 0 and Sell : 1 --> Easier comparisons
data = pd.get_dummies(df,prefix=[''], columns = ['bors'], drop_first=False)
data.drop('_S', axis=1, inplace=True)
```

```python
data.rename(columns = {"_B":"isbuy"},inplace = True)
```

```python
data = data.fillna(0)
```

```python
## label encoder to convert date into an integer representation

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data[['date']] = data[['date']].apply(le.fit_transform)
```

```python
data.head()
```

```python
data.to_csv(r'numpydatabook.txt', header=None, index=None, sep=' ', mode='a')
```
